# MPC cost sensitivity
Explore pressure vs energy trade-offs as energy weight varies.

```python
import numpy as np
import matplotlib.pyplot as plt
import os
np.random.seed(0)
weights = np.linspace(0, 1, 21)
pressure = 40 + 10*(1-weights) + np.random.normal(scale=0.5, size=weights.shape)
energy = 20 + 15*weights + np.random.normal(scale=0.5, size=weights.shape)
plt.figure(figsize=(4,3))
plt.scatter(energy, pressure)
plt.xlabel('energy usage')
plt.ylabel('pressure penalty')
plt.tight_layout()
os.makedirs('../plots', exist_ok=True)
plt.savefig('../plots/mpc_cost_sensitivity.png', bbox_inches='tight')
plt.show()
print('Saved figure to ../plots/mpc_cost_sensitivity.png')
```
